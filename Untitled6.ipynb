```python
import emoji

def give_emoji_free_text( text):
    allchars = [str for str in text]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.split() if not any(i in str for i in emoji_list)])

    return clean_text


fileexcel=open("F:\\Event\\Events.xlsx",'a')
workbook = xlsxwriter.Workbook("F:\\Event\\Events.xlsx")
worksheet=workbook.add_worksheet()
worksheet.write('A1','Event Name')
worksheet.write('B1','Event Address')
worksheet.write('C1','Event Date')
worksheet.write('D1','Event Time')
worksheet.write('E1','Event kind')
worksheet.write('F1','Event URL')
worksheet.write('G1','Ticket Website') 
worksheet.write('H1','host link ')
worksheet.write('I1','host link ')  
worksheet.write('J1','host link ')
worksheet.write('K1','host name')  
worksheet.write('L1','host name2')
worksheet.write('M1','host name3')  
worksheet.write('N1','host name4')
worksheet.write('O1','host name5')  
worksheet.write('P1','host name6')
worksheet.write('Q1','Details')


index=2
details_list=list()
delays = [7, 4, 6, 13, 10]


lens=int (len(ids)/10)
start=0
end=10
for lb in range(0,lens):
    for link in range(start,end):
        delay = np.random.choice(delays)
        sleep(delay)
        driver.get(event_links[link])
    
    
    
        #name
        name_=driver.find_elements_by_xpath("//*[contains(@class,'_31y6')]")
        name=name_[0].text
        worksheet.write('A'+str(index),name)
    
    
    
    
        #date
        date_=driver.find_elements_by_xpath("//*[contains(@class,'_mie _2ob _38nm _38nq')]")
        date=date_[0].get_attribute('title')
        worksheet.write('C'+str(index),date)
    
    
        #kind
        comment_=driver.find_elements_by_xpath("//*[contains(@class,'_31y7')]")
        comment=comment_[0].text
        worksheet.write('E'+str(index),comment)
    
    
    
        #Event URL
        link2=(event_links[link]).replace('m.','www.',1)
        worksheet.write('F'+str(index),link2)
    
    
        #host
        source_=driver.find_elements_by_xpath("//*[contains(@class,'_4e81')]")
        for linx in source_:
            source=(linx.get_attribute('href')).replace('m.','www.',1)
            if(source_.index(linx)==0):
                worksheet.write('H'+str(index),source)
            if(source_.index(linx)==1):
                worksheet.write('I'+str(index),source)
            if(source_.index(linx)==2):
                worksheet.write('J'+str(index),source)

        
        
        
        #time
        data_=driver.find_elements_by_xpath("//*[contains(@class,'_52je _52jb _52jg')]")
        data=driver.find_elements_by_xpath("//*[contains(@class,'_52jc _52j9 _52jg')]")
        date2_=(data_[0].text)+" --( "+(data[0].text)+" )"
        if (date2_==((data_[0].text)+' --(  )')):
            date2_=data_[0].text
        date2=(date2_.split(" at ",1))[-1]
        worksheet.write('D'+str(index),date2)
    
    
    
    
        #address
        address=(data_[1].text)+" --( "+(data[1].text)+" )"
        if (address==((data_[1].text)+' --(  )')):
            address=data_[1].text
        worksheet.write('B'+str(index),address)
    
    
        #ticket
        try:
            ticket_=driver.find_element_by_xpath("//*[contains(@class,'_20zd')]")
            ticket=ticket_.text
        except:        
            ticket=' '
        worksheet.write('G'+str(index),ticket)
    
    
    
        #details
        try:
            driver1 = driver.find_element_by_xpath("//*[contains(@class,'_52ja _2pi9 _2pip _2s23')]")
            element = driver1.find_elements_by_link_text("More")
            element[-1].click()
        except:
            pass
        details1_ = driver.find_element_by_xpath("//*[contains(@class,'_52ja _2pi9 _2pip _2s23')]")
        #details1=(((details1_.text).replace("\n","  ")).replace("… More",".....  "))
        sleep(3)
        text1=(details1_.text).replace("\n","(n)")
        text2=(text1).encode('utf-8').decode('ascii', 'ignore')
        text3=give_emoji_free_text(text=text2)
        worksheet.write('Q'+str(index),give_emoji_free_text(text=text2.replace("(n)","----")))
        details_list.append(give_emoji_free_text(text=text2.replace("(n)","----")))

    
    
    
    
    
        #host name
        hosted= driver.find_elements_by_xpath("//*[contains(@class,'_9_7 _2rgt _1j-g _2rgt _86-3 _2rgt _1j-g _2rgt')]") 
        for lin in hosted:
            if(hosted.index(lin)==0):
                worksheet.write('K'+str(index),lin.text)
            if(hosted.index(lin)==1):
                worksheet.write('L'+str(index),lin.text)
            if(hosted.index(lin)==2):
                worksheet.write('M'+str(index),lin.text)
            if(hosted.index(lin)==3):
                worksheet.write('N'+str(index),lin.text)
            if(hosted.index(lin)==4):
                worksheet.write('O'+str(index),lin.text)
            if(hosted.index(lin)==5):
                worksheet.write('P'+str(index),lin.text)
        print("-------------------------------------------------------------------"+str(index-1))
        index=index+1
    start=end
    end=end+10
    print('loop = '+ str(lb+1))
    sleep(10)
workbook.close()

print ("Done") 
input('Press anything to quit') 
driver.quit() 
print("Finished") 
```
